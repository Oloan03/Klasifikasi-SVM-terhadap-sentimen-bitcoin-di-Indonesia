```python
import pandas as pd
```

```python
df = pd.read_csv('../preprocessing/clean_text.csv')
df
```

```python
df.drop(columns=['full_text', 'case_folding', 'tokens', 'normalized', 'stopword_removal'], inplace=True)
df
```

```python
#Labeling

#set lexicon
pos_inset = pd.read_csv('../../Inset/positive.tsv', sep='\t', header=None, names=['word', 'weight'])
neg_inset = pd.read_csv('../../Inset/negative.tsv', sep='\t', header=None, names=['word', 'weight'])
sentis = set(pd.read_csv('../../sentistrength_id/sentiwordsid.csv', header=None)[0])
import pandas as pd
import ast


# Pastikan kolom 'stemming' diubah dari string ke list
df['stemming'] = df['stemming'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Baca leksikon Inset
# Format tsv: kata\tbobot (atau hanya kata saja)
# Kita akan asumsikan format: kata\tbobot

# Buat kamus leksikon
positive_lexicon = set(pos_inset['word'].str.lower().tolist())
negative_lexicon = set(neg_inset['word'].str.lower().tolist())

print(f"Jumlah kata positif: {len(positive_lexicon)}")
print(f"Jumlah kata negatif: {len(negative_lexicon)}")

def label_sentiment(tokens):
    """
    Label sentimen berdasarkan kemunculan kata dalam leksikon Inset.
    Hanya dua kelas: positif dan negatif.
    Jika tidak ada kata yang cocok, beri label 'netral'.
    """
    if not isinstance(tokens, list):
        return 'netral'
    
    # Hitung kemunculan kata positif dan negatif
    pos_count = sum(1 for word in tokens if word.lower() in positive_lexicon)
    neg_count = sum(1 for word in tokens if word.lower() in negative_lexicon)
    score = pos_count - neg_count
    # Hanya dua kelas: positif dan negatif
    # Jika ada kata positif dan tidak ada kata negatif → positif
    # Jika ada kata negatif dan tidak ada kata positif → negatif
    # Jika ada keduanya atau tidak ada sama sekali → netral (tapi kita beri label negatif sebagai default)
    
    if pos_count > 0 and neg_count == 0:
        label = 'positif'
    elif neg_count > 0 and pos_count == 0:
        label = 'negatif'
    elif score > 0:
        label = 'positif'
    else:
        # Jika sama atau tidak ada yang cocok, default ke 'negatif'
        # Atau bisa diubah ke 'netral' jika ingin tiga kelas
        label = 'negatif'
    
    return score, label

# Terapkan pelabelan
df[['score','sentiment']] = df['stemming'].apply(lambda x: pd.Series(label_sentiment(x)))

# Tampilkan distribusi
sentiment_dist = df['sentiment'].value_counts()
print("\nDistribusi Sentimen:")
print(sentiment_dist)

# Tampilkan beberapa contoh
print("\nContoh hasil pelabelan:")
for i, row in df.head(10).iterrows():
    print(f"Stemming: {row['stemming'][:10]}...")
    print(f"Sentimen: {row['sentiment']}")
    print("-" * 80)

#df['sentiment'] = df['stemming'].apply(lambda x: pd.Series(label_sentiment_weighted(x)))
#df[['sentis_score','label_sentis']] = df['stemming'].apply(lambda x: pd.Series(labeling_sentis(x)))

df
```

```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('../preprocessing/clean_text.csv')
df
```

```python
df.drop(columns=['full_text', 'case_folding', 'tokens', 'normalized', 'stopword_removal'], inplace=True)
df
```

```python
df.drop_duplicates(subset=['stemming'], inplace=True)
df
```

```python
#Labeling

#set lexicon
pos_inset = pd.read_csv('../../Inset/positive.tsv', sep='\t', header=None, names=['word', 'weight'])
neg_inset = pd.read_csv('../../Inset/negative.tsv', sep='\t', header=None, names=['word', 'weight'])
sentis = pd.read_csv('../../sentistrength_id/sentiwordsid.csv', header=None, names=['word', 'weight'])
sentis = set(sentis['word'].str.lower())


# Pastikan kolom 'stemming' diubah dari string ke list
df['stemming'] = df['stemming'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Baca leksikon Inset
# Format tsv: kata\tbobot (atau hanya kata saja)
# Kita akan asumsikan format: kata\tbobot

# Buat kamus leksikon
positive_lexicon = set(pos_inset['word'].str.lower().tolist())
negative_lexicon = set(neg_inset['word'].str.lower().tolist())

print(f"Jumlah kata positif: {len(positive_lexicon)}")
print(f"Jumlah kata negatif: {len(negative_lexicon)}")

def inset_label_sentiment(tokens):
    """
    Label sentimen berdasarkan kemunculan kata dalam leksikon Inset.
    Hanya dua kelas: positif dan negatif.
    Jika tidak ada kata yang cocok, beri label 'netral'.
    """
    if not isinstance(tokens, list):
        return 'netral'
    
    #Hitung pola kemunculan kata
    pos_count = sum(word.lower() in positive_lexicon for word in tokens)
    neg_count = sum(word.lower() in negative_lexicon for word in tokens)
    
    score = pos_count - neg_count
    #Tentukan label berdasarkan bobot
    if score < 0:
        label = 'negatif'
    elif score > 0:
        label = 'positif'
    else:
        # Jika sama atau tidak ada yang cocok, default ke 'negatif'
        # Atau bisa diubah ke 'netral' jika ingin tiga kelas
        label = 'netral'
    
    return score, label

def determine_sentiment(text):
    if not isinstance(text, str):
        return 0, 'netral'
    
    # Hitung skor sentimen dengan bobot
    pos_words = sum(word in sentis for word in text)
    neg_words = sum(word in sentis for word in text)
    
    # Hitung skor sentimen dengan mempertimbangkan jumlah kata
    total_words = pos_words + neg_words
    if total_words == 0:
        sentiment_score = 0
    else:
        sentiment_score = (pos_words - neg_words) / total_words
    
    # Tentukan threshold untuk klasifikasi
    if sentiment_score > 0:  # Threshold positif
        sentiment = 'positif'
    elif sentiment_score < 0:  # Threshold negatif
        sentiment = 'negatif'
    else:  # Zona netral, gunakan pendekatan probabilistik
        sentiment = 'netral'
    
    return sentiment_score, sentiment


# Terapkan pelabelan
df[['inset_score','inset_sentiment']] = df['stemming'].apply(lambda x: pd.Series(inset_label_sentiment(x)))
df[['sentis_score','sentis_sentiment']] = df['stemming'].apply(lambda x: pd.Series(determine_sentiment(x)))

# Tampilkan distribusi
inset_sentiment_dist = df['inset_sentiment'].value_counts()
sentis_sentiment_dist = df['sentis_sentiment'].value_counts()
print("\nDistribusi Sentimen:")
print(inset_sentiment_dist)
print("\nDistribusi Sentimen Sentis:")
print(sentis_sentiment_dist)

# Tampilkan beberapa contoh
print("\nContoh hasil pelabelan inset:")
for i, row in df.head(10).iterrows():
    print(f"Stemming: {row['stemming'][:10]}...")
    print(f"Score: {row['inset_score']}")
    print(f"Sentimen: {row['inset_sentiment']}")
    print("-" * 80)
    
print("\nContoh hasil pelabelan sentis:")
for i, row in df.head(10).iterrows():
    print(f"Stemming: {row['stemming'][:10]}...")
    print(f"Score: {row['sentis_score']}")
    print(f"Sentimen: {row['sentis_sentiment']}")
    print("-" * 80)

#df['sentiment'] = df['stemming'].apply(lambda x: pd.Series(label_sentiment_weighted(x)))
#df[['sentis_score','label_sentis']] = df['stemming'].apply(lambda x: pd.Series(labeling_sentis(x)))

df
```
